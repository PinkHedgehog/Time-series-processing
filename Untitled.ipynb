```python
crime_df = pd.read_csv('crime.csv', parse_dates=['month'], index_col=['month'], dayfirst=True)
crime_df.describe()
```

```python
plt.plot(crime_df.index, crime_df['Terrorism'], label='Terrorism')
plt.plot(crime_df.index, crime_df['Extremism'], label='Extremism')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
description_of_crimes = crime_df.describe()
```

```python
crime_types = list(crime_df.keys())
crime_types_sorted_by_mean = sorted(crime_types, key=lambda c: description_of_crimes[c]['mean'])
```

```python
train = crime_df[crime_df.index < pd.to_datetime("2016-11-01", format='%Y-%m-%d')]
test = crime_df[crime_df.index > pd.to_datetime("2016-11-01", format='%Y-%m-%d')]
plt.plot(train['Extremism'], color = "black")
plt.plot(test['Extremism'], color = "red")
plt.ylabel('# of cirmes')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.title(f"Train/Test split for {'Extremism'}")
plt.show()
```

```python
def visualize(crime_types, train_df, test_df):
    """PLot images in one row."""
    n = len(crime_types)
    plt.figure(figsize=(32, 24))
    
    for i, ct in enumerate(crime_types):
        plt.subplot(n % 5 + 1, n // 5 + 1, i + 1)
        #plt.ylabel('# of cirmes')
        #plt.xlabel('Date')
        plt.xticks(rotation=45)

        plt.title(ct)
        plt.plot(train_df[ct], color = "blue")
        plt.plot(test_df[ct], color = "orange")
    plt.show()
```

```python
visualize(crime_types_sorted_by_mean[:-1], train, test)
```

```python
y = crime_df['Extremism']
```

```python
ARMAmodel = SARIMAX(y, order = (1, 0, 1))
ARMAmodel = ARMAmodel.fit()
```

```python
y_pred = ARMAmodel.get_forecast(len(test.index))
y_pred_df = y_pred.conf_int(alpha = 0.05)
print(type(y_pred_df))
y_pred_df["Predictions"] = ARMAmodel.predict(start = y_pred_df.index[0], end = y_pred_df.index[-1])
y_pred_df.index = test.index
y_pred_out = y_pred_df["Predictions"] 
```

```python
plt.plot(y_pred_out, color='green', label = 'Predictions - Extremism')
plt.xticks(rotation=45)
plt.legend()
```
